```python

import pandas as pd

# 数据加载
train = pd.read_csv('./train_set.csv', sep='\t')
test = pd.read_csv('./test_a.csv', sep='\t')
train
```

```python
# 将训练集和测试集合并
df_all = pd.concat([train, test])
df_all
```

```python
## 提取TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
# 抽取ngram_range = (1, 3)
vec = TfidfVectorizer(ngram_range=(1, 3), min_df=0.1, max_df=0.8)
# 提取tfidf
text_features = vec.fit_transform(df_all['text'])
text_features
```

```python
text_features.shape
```

```python
df = pd.DataFrame(text_features.toarray(), columns=vec.get_feature_names())
df
```

```python
df.to_pickle('./df.pkl')
```

```python
df = pd.read_pickle('./df.pkl')
```

```python
# 还原训练集，测试集
df_train = df[df.index < len(train)]
df_test = df[df.index >= len(train)]
df_test.index = range(len(df_test))
df_test
```

```python
# 构造text_length
train['text_len'] = train['text'].apply(lambda x: len(x.split(' ')))
test['text_len'] = test['text'].apply(lambda x: len(x.split(' ')))
test
```

```python

# 计算唯一值个数
def f(x):
    temp = set(x.split(' '))
    return len(temp)
    
train['text_nunique'] = train['text'].apply(f)
train
```

```python
test['text_nunique'] = test['text'].apply(f)
test
```

```python
# 去掉text列
train.drop(['text'], axis=1, inplace=True)
test.drop(['text'], axis=1, inplace=True)
train
```

```python
# 拼接到df_train
train = train.merge(df_train, left_index=True, right_index=True)
test = test.merge(df_test, left_index=True, right_index=True)
train
```

```python

train.to_pickle('./train.pkl')
test.to_pickle('./test.pkl')
```

```python
from lightgbm.sklearn import LGBMClassifier
clf = LGBMClassifier(num_leaves=2**5-1, reg_alpha=0.25, reg_lambda=0.25, objective='multiclass', max_depth=-1, learning_rate=0.005, min_child_samples=3, random_state=2021, n_estimators=2000, subsample=1, colsample_bytree=1)
clf = clf.fit(train.drop('label', axis=1), train['label'])
```

```python
y_pred = clf.predict(test)
y_pred = pd.DataFrame(y_pred, columns=['label'])
```

```python
y_pred.to_csv('./version2_lgb_2000.csv', index=False)
y_pred
```
